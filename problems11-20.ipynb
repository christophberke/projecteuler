### Problem 12:

```julia
function count_divisor(n)
	if n <= 2
		return n 
	end 

	count_divs = 2 # Start with two: 1 and n are always divisors.
	for i in 2:sqrt(n)
		if rem(n, i) == 0 
			count_divs += 2  # +2 because i and n/i are divisors.
		end 
	end 

	return count_divs 
end 

n = 1
i = 1
n_div = 1

while n_div < 500 
# for _ in 1:10
	global i += 1
	global n = n + i
	global n_div = count_divisor(n) 
	println(n, ",  ", n_div)
end 
```

```julia
update_step(n) = iseven(n) ? (return div(n,2)) : (return 3*n + 1)

i_max, count_max = 1, 1
for i in 1:1_000_000

	n = i
	count = 0
	while n > 1
		n = update_step(n)
		count += 1
	end 

	if count > count_max
		global count_max = count 
		global i_max = i
	end

end

i_max
```
